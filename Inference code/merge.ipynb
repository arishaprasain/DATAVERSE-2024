**Importing pandas for merge**

```python
import pandas as pd
```

**Reading the csv files**

```python
bbox = pd.read_csv('/content/drive/MyDrive/DATAVERSE 2024/boundingbox.csv')
```

```python
invoice_info = pd.read_csv('/content/drive/MyDrive/DATAVERSE 2024/testoutput.csv')
```

```python
bbox.head(3)
```

```python
invoice_info.head()
```

**We can see that above two dataframes can be merged on the same column 'title' which is present in both dataframes to obtain desired dataframe**

Merging

```python
merged_df = pd.merge(invoice_info, bbox , on = 'title')
```

```python
merged_df.head(5)
```

**Rearranging the dataframe columns to match the file provided**

```python
column_order = ['title', 'invoice_number', 'invoice_number_bbox', 'issue_date', 'issue_date_bbox',
                'total','total_bbox', 'table', 'table_bbox'
]
```

```python
final_df = merged_df[column_order]
```

```python
final_df
```

**Final dataframe obtained in the required format!**

**Saving the output fie as csv ...**

```python
csv_file_path = '/content/drive/MyDrive/DATAVERSE 2024/predictions.csv'
final_df.to_csv(csv_file_path, index=False)
```
